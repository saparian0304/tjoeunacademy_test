```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from datetime import datetime
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
```

```python
url = 'https://www.youtube.com/channel/UCzlTLjXTQrzCnx0CO5sgjdQ/videos'
br = webdriver.Chrome('chromedriver')
br.get(url)

# 페이지 높이 반환
scroll_pane_height = br.execute_script('return document.documentElement.scrollHeight')
```

```python
import time
# 스크롤바 내리기
while True:
    br.execute_script('window.scrollTo(0,document.documentElement.scrollHeight)')
    time.sleep(1)
    new_scroll_pane_height = br.execute_script('return document.documentElement.scrollHeight')
    if scroll_pane_height == new_scroll_pane_height : 
        break
    scroll_pane_height = new_scroll_pane_height
    
```

```python
html = br.page_source
soup = BeautifulSoup(html, 'html.parser')
tag_list = soup.select('#video-title-link')
    
for tag in tag_list:
    print(tag.get('href'), tag.text)
```
