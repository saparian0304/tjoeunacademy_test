```python
# 최종
import pandas as pd
import requests
import urllib.request as req
from bs4 import BeautifulSoup
lst = []
for m in range(1,34):
    url='https://www.animal.go.kr/front/awtis/institution/institutionList.do?totalCount=319&pageSize=10&menuNo=1000000059&&page={}'.format(m)

    html = requests.get(url)
    html.encoding='utf-8' # requests한글 깨짐
    soup = BeautifulSoup(html.text, 'html.parser')
    
    for t in range(1,10):
        try:
            lst_ = []
            title = soup.select_one('#searchList > div.boardList2 > table > tbody > tr:nth-child({}) > td:nth-child(2) > a'.format(t+1))
            tel = soup.select_one('#searchList > div.boardList2 > table > tbody > tr:nth-child({}) > td:nth-child(3) > a'.format(t+1))
#             add = soup.select_one('#searchList > div.boardList2 > table > tbody > tr:nth-child({}) > td:nth-child(4) > a'.format(t+1))
            content = soup.select_one('#searchList > div.boardList2 > table > tbody > tr:nth-child({}) > td:last-child'.format(t+1))
#             pet = '상호명 :   {} 전화번호 :   {}     주소 :   {}'.format(title.string, tel.string, content.string)
            pet = '{},   {},     {}'.format(title.string, tel.string, content.string)
#             print(pet)
            lst_ = [title.string, tel.string, content.string]
            lst.append(lst_)
        except Exception as e:
            print(e)
            pass
 
        
df = pd.DataFrame(lst)
df
```
