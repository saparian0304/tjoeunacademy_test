### BeautifulSoup
* html, xml에서 원하는 데이터를 추출
* 다운로드 안됨
* 아나콘다에 내장되어있음

```python
from bs4 import BeautifulSoup
```

```python
html = '''
<html>
    <body>
        <h1>스크래핑</h1>
        <div>크롤링</div>
        <span>HTML</span>
    </body>
</html>
'''

# 응답받은 내용을 파싱(해석, 번역)
soup = BeautifulSoup(html)

# 원하는 데이터 추출(구조적으로)
text1 = soup.html.body.h1
text2 = soup.html.body.div
text3 = soup.html.body.span

print(text1.string)
print(text2.string)
print(text3.string)
```

## 특정 부분만 선택

```python
html = '''
<html>
    <body>
        <h1 id='title'>스크래핑</h1>
        <div class='content'>크롤링</div>
        <span>HTML</span>
    </body>
</html>
'''
# find() : 하나만 추출
soup = BeautifulSoup(html)
title = soup.find(id='title')

print(title.string)
content = soup.find(class_='content')
print(content.string)
```

```python
# find_all() : 여러개 추출
html = '''
<html>
    <body>
        <a href='http://www.naver.com'>네이버</a>
        <a href='http://www.daum.net'>다음</a>
        <a href='http://www.google.com'>구글</a>
    </body>
</html>
'''
soup = BeautifulSoup(html)
datas = soup.find_all('a')
for d in datas:
    print(d.string, d.attrs['href'])
```

### CSS 선택자

```python
url = 'https://news.daum.net/'
res = req.urlopen(url)
soup = BeautifulSoup(res, 'html.parser')
# 추출
# 하나만 추출 (개발자도구 -> copy selector 이용)
text = soup.select_one('body > div.container-doc > main > section > div > div.content-article > div.box_g.box_news_issue > ul > li:nth-child(1) > div > div > strong > a')  
# print(text.string.strip())

# 여러개 추출
text = soup.select('body > div.container-doc > main > section > div > div.content-article > div.box_g.box_news_issue > ul > li > div > div > strong > a') 
# print(text)
for t in text:
    print(t.string.strip(), t.attrs['href'])
```

### 네이버 환율 데이터 추출  
### 원/달러, 원/엔
네이버 > 증권 > 시장지표  
https://finance.naver.com/marketindex/


```python
# 달러환율 : 0000
# 엔화환율 : 0000
url = 'https://finance.naver.com/marketindex/'
res = req.urlopen(url)
soup = BeautifulSoup(res, 'html.parser')
text = soup.select('#exchangeList > li > a.head > div > span.value')

for t in text[:2]:
    print(t.string)
print('달러환율 : {}원'.format(text[0].string))
print('엔화환율 : {}원'.format(text[1].string))
```

```python
url = 'https://finance.naver.com/marketindex/exchangeList.naver'
res = req.urlopen(url)
soup = BeautifulSoup(res, 'html.parser')
text = soup.select('body > div > table > tbody > tr > td.sale')
title = soup.select('body > div > table > tbody > tr > td.tit')

for t in range(len(title)):
    print(f'{title[t].string.strip()} : {text[t].string} 원')
```

```python
url = 'https://finance.naver.com/marketindex/exchangeList.naver'
res = req.urlopen(url)
soup = BeautifulSoup(res, 'html.parser')
text = soup.select('body > div > table > tbody > tr')
for t in text:
    string = t.get_text(strip=True, separator=',,').split(',,')
    print(f'{string[0]} : {string[1]}원')
    for a in t:
        print(a.string.strip())
```

```python
url = 'https://finance.naver.com/marketindex/exchangeList.naver'
res = req.urlopen(url)
soup = BeautifulSoup(res, 'html.parser')
text = soup.select('tbody > tr')
for t in text:
    print(t.find(class_='tit').string.strip(), t.find(class_='sale').string.strip(), "원")
```
