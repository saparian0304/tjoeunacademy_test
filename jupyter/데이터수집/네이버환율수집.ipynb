```python
url = 'https://finance.naver.com/marketindex/'
import requests
```

```python
res = requests.get(url)
res.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# 달러
dollar = soup.select_one('#exchangeList > li > a.head.usd > div > span.value')
dollar = float(dollar.string.replace(',',''))
```

```python
# 엔화
yen = soup.select_one('#exchangeList > li > a.head.jpy > div > span.value')
yen = float(yen.string.replace(',',''))
```

```python
dollar, yen
```

```python
import pymysql
conn = pymysql.connect(host='localhost', user='testuser', password='test1234', db='project', charset='utf8')
cur = conn.cursor()
```

```python
cur.execute('INSERT INTO exchange (dollar, yen, date) VALUES(%s, %s, now())', (dollar, yen))
conn.commit()
```

```python
cur.close()
conn.close()
```
