```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
url = 'http://www.weatheri.co.kr/bygone/pastDB_tmp.php?jijum_id=108&start=2021-1-01&end=2021-12-30'
br = webdriver.Chrome('chromedriver')
br.get(url)

from bs4 import BeautifulSoup
html = br.page_source
soup = BeautifulSoup(html, 'html.parser')
row_count = soup.select('body > table > tbody > tr > td > form > table:nth-child(2) > tbody > tr > td > table > tbody > tr')
dates = []
avgs = []
lows = []
highs = []
for m in range(12):
    for i in range(2, len(row_count)+1):

        date = soup.select_one(f'body > table table:nth-child(2) table > tbody > tr:nth-child({i}) > td:nth-child(1)').get_text()

        avg_temp = soup.select_one(f'body > table table:nth-child(2) table > tbody > tr:nth-child({i}) > td:nth-child(2)').string

        high_temp = soup.select_one(f'body > table table:nth-child(2) table > tbody > tr:nth-child({i}) > td:nth-child(3)').string

        low_temp = soup.select_one(f'body > table table:nth-child(2) table > tbody > tr:nth-child({i}) > td:nth-child(4)').string
        dates.append(date)
        avgs.append(avg_temp)
        lows.append(low_temp)
        highs.append(high_temp)
        print(date, avg_temp, high_temp, low_temp)
    br.find_element(By.CSS_SELECTOR, 'body > table > tbody > tr > td > form > table:nth-child(1) > tbody > tr > td:nth-child(1) > select').send_keys(Keys.ARROW_DOWN)

    br.find_element(By.CSS_SELECTOR, 'body > table > tbody > tr > td > form > table:nth-child(1) > tbody > tr > td:nth-child(2) > img').click()
    html = br.page_source
    soup = BeautifulSoup(html, 'html.parser')
    row_count = soup.select('body > table > tbody > tr > td > form > table:nth-child(2) > tbody > tr > td > table > tbody > tr')
    
br.quit()
```

```python
for i in range(len(dates)):
    print(dates[i], avgs[i], lows[i], highs[i])
```
