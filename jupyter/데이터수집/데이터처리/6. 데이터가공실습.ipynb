```python
# 온라인_판매기록.csv 불러오기
import pandas as pd
df = pd.read_csv('../data/온라인_판매기록.csv', encoding='MS949')
```

```python
# 피벗테이블로 행:제품, 열:쇼핑몰 유형, 값: 판매금액
# 평균 소수점 2자리로 반올림
import numpy as np
# 첫번째 방법
pd.pivot_table(df, index='제품', columns='쇼핑몰 유형', values='판매금액').round(2)

# 두번째 방법 - 함수 정의해서 aggfunc에 넣기
def my_mean(v):
    return round(np.mean(v), 2)
pd.pivot_table(df, index='제품', columns='쇼핑몰 유형', values='판매금액', aggfunc=my_mean)

# 세번째 방법 - lambda 사용
pd.pivot_table(df, index='제품', columns='쇼핑몰 유형', values='판매금액', aggfunc=lambda v:round(np.mean(v), 2))
```

```python
# 쇼핑몰 유형별 판매금액 평균
df.groupby('쇼핑몰 유형').mean()['판매금액'].round(2)
df.groupby('쇼핑몰 유형')['판매금액'].mean().round(2)
```

```python
# 쇼핑몰 유형별 판매금액, 수량 평균
df.groupby('쇼핑몰 유형')[['판매금액', '수량']].mean().round(2)
```

```python
# 쇼핑몰 유형별 판매금액, 수량 평균, 최대값
df.groupby('쇼핑몰 유형')[['판매금액', '수량']].agg(['mean', 'max'])
```

```python
# 쇼핑몰 유형별 판매금액, 수량 최대-최소 차이
def sub(v):
    return max(v) - min(v)
df.groupby('쇼핑몰 유형')[['판매금액', '수량']].agg(['max', 'min', sub])
df.groupby('쇼핑몰 유형')[['판매금액', '수량']].agg(lambda v : max(v) - min(v))
```
