* concat : row나 column을 기준으로 단순하게 이어붙이기
* merge : 특정 값을 기준으로 병합

```python
'''
pd.merge(left, right, on='기준컬럼', how='left|right|inner|outer')
'''
df3 = df2.drop([1, 3, 5, 7, 9])
df3
```

```python
df3 = df3.reset_index(drop=True)
```

```python
df3
```

```python
pd.concat([df, df3], axis=1) # 행 인덱스명 기준으로 붙기때문에 문제가 있음
```

```python
pd.merge(df, df3, on='이름', how='left')
```

```python
pd.merge(df, df3, on='이름', how='right')
```

```python
df3.columns = ['성명', '연봉', '가족수']
df3
```

```python
pd.merge(df, df3, on='이름', how='left') # df3에 이름이라는 컬럼명이 없으므로 에러
```

```python
pd.merge(df, df3, left_on='이름', right_on='성명', how='left')
```
