### 피벗테이블
- 엑셀 피벗테이블
- 컬럼 중에서 두개를 골라서 행 인덱스, 열 인덱스로 사용해서 데이터를 펼쳐놓은 것 
- index : 행, cloumns : 열, values : 값

```python
pd.pivot_table(df, index='소속사', columns='혈액형', values='키') # 기본값은 평균
```

```python
def avg(row):
    total = 0
    for r in row:
        total += r
    return total / len(row)
```

### 라벨인코딩(Label Encoding)


```python
# 성별 남자:1, 여자:2
df.loc[df['성별']=='남자', '성별'] = 1
df.loc[df['성별']=='여자', '성별'] = 2
```

```python
df
```

```python
df = pd.read_csv('../data/korean-idol.csv')
# apply 함수를 사용해서 처리
def gender(x):
    if x['성별'] == '남자':
        return 1
    elif x['성별'] == '여자':
        return 2
```

```python
df['성별'] = df.apply(gender, axis=1)
```

```python
df
```

```python
df = pd.read_csv('../data/korean-idol.csv')
```

```python
df['성별'].apply(lambda x:1 if x=='남자' else 2)
```

```python
# map (값 매핑)
gender_map = {'남자':1, '여자':2}
df['성별'].map(gender_map)
```
