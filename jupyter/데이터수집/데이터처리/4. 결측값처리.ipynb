```python
import pandas as pd
df = pd.read_csv('../data/korean-idol.csv')
df
```

```python
df.info()
```

```python
# 결측치 확인
df.isnull().sum()
```

### fillna()
- na값을 fill(채워주는) 함수

```python
df['키'].fillna(0)
```

```python
df['키']
```

```python
# df['키'] = df['키'].fillna(0)
# 또는 
df['키'].fillna(0, inplace=True)
df['키']
```

```python
# 결측치를 치환할 때 자주 사용되는 대표값
# 평균, 중앙값(수치형) + 최빈값(범주형)
df['키'].mean(), df['키'].median()
```

```python
df['키'].mode() # 시리즈 타입으로 나옴
df['키'].mode()[0]
```

```python
df['키'].fillna(df['키'].mode()[0])
```

```python
df.info()
```

### 결측값이 있는 행을 제거

```python
df.dropna() # 하나라도 결측치가 있으면 그 행을 삭제
```

```python
df.dropna(axis=1) # 하나라도 결측치가 있으면 그 행을 삭제
```

```python
# how 옵션 : any: 하나라도 있으면 삭제, all: 전부 결측치면 삭제
df.dropna(axis=0, how='all')
```

```python
import numpy as np
df.loc[15] = np.nan # 15 인덱스에  nan값 추가
df
```

```python
df.dropna(axis=0, how='all')
```

### 컬럼의 중복값을 제거

```python
df['키'].drop_duplicates()
```

```python
# keep 옵션 first / last
df['키'].drop_duplicates(keep='last')
```

```python
# 행단위 제거
df.drop_duplicates('키')
```

```python
df.drop('키', axis=1)
```
