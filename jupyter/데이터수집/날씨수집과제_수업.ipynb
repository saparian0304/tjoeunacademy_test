```python
import requests
from bs4 import BeautifulSoup

for m in range(1, 13):
    url = 'https://www.weatheri.co.kr/bygone/pastDB_tmp.php?jijum_id=108&start=2021-{}-01&end=2021-12-31'.format(m)
    html = requests.get(url)
    html.encoding='utf-8'
#     print(html.text)
    soup = BeautifulSoup(html.text, 'html.parser')
    
    for d in range(1, 32):
        try:
            avg = soup.select_one('table table:nth-of-type(2) table > tr:nth-of-type({}) > td:nth-of-type(2) > font'.format(d+1))
            high = soup.select_one('table table:nth-of-type(2) table > tr:nth-of-type({}) > td:nth-of-type(3) > font'.format(d+1))
            low = soup.select_one('table table:nth-of-type(2) table > tr:nth-of-type({}) > td:nth-of-type(4) > font'.format(d+1))
            print('{}월 {}일 평균기온 : {}, 최고기온: {}, 최저기온 : {}'.format(m, d, avg.string, high.string, low.string))
        except:
            pass
```
