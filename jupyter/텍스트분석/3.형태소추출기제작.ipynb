```python
# 전처리 함수
def preprocessing(txt):
    import re
    txt = re.sub('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]', '', txt)
    sw = set()
    with open('stopwords-ko.txt', encoding='utf-8') as f:
        for word in f:
            sw.add(word.replace('\n', ''))
    tokens = []
    from eunjeon import Mecab
    mecab = Mecab()
    for t in mecab.morphs(txt):
        if t not in sw and len(t) > 1:
            tokens.append(t)    
    return tokens
```
