### token(토큰)
* 문자열(문장, 문서)를 이루는 최소한의 단위 -> 단어
* 문자열을 토큰으로 분리하는 작업을 tokenizing

```python
from nltk.tokenize import sent_tokenize
print(sent_tokenize(data[:1000])[0])
```

```python
from nltk.tokenize import word_tokenize
print(word_tokenize(data[:1000]))
```

```python
# 정규식 토크나이징
from nltk.tokenize import RegexpTokenizer
ret = RegexpTokenizer('[\w]+')  # 문자만 토큰화하기 위한 정규식
ret.tokenize(data[:1000])
```

### morph (형태소)
- 의미있는 가장 작은 단위
- 보통 토큰 단위
- 품사, 접두사, 접미사 등...

### 품사(pos, part-of-speech)
- 문법적인 의미/뜻으로 구분
- 동사, 명사...

```python
from nltk.tag import pos_tag
sent = 'The family of Dashwood had long been settled in Sussex.'
pos_list = pos_tag(word_tokenize(sent))
pos_list # 각 토큰에 품사를 tag 해줌, 'NNP' : 고유명사
```

```python
# 명사만 추출
[p[0] for p in pos_list if p[1] == 'NN']
```

```python
# Text (단어 단위 빈도)
from nltk import Text
txt = Text(ret.tokenize(data))
txt[:100]
```

```python
# 빈도 그래프
txt.plot(20)
```

```python
# FreqDist 클래스
# 토큰의 사용빈도
# Text클래스의 vocab() 메서드로 추출
fd = txt.vocab()
type(fd)
```

```python
from nltk import FreqDist
tokens = pos_tag(ret.tokenize(data))
tokens
```

```python
# NNP만 추출
nnp = [p[0] for p in tokens if p[1] == 'NNP']
nnp[:10]
```

```python
fd = FreqDist(nnp)
# 전체 단어수, 해당단어출현횟수, 확률
fd.N(),  fd['Dashwood'], fd.freq('Dashwood')
```

```python
252/6530
```

```python
# 가장 많이 출현횟수가 높은 단어
fd.most_common(10)
```
