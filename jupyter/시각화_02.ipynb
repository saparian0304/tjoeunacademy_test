## seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 한글처리
from matplotlib import font_manager, rc
# fm = font_manager.FontProperties(fname='C:\Windows\Fonts\HMFMPYUN.TTF')
# rc('font', family=fm.get_name())
rc('font', family='D2Coding')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 바이올린
sns.violinplot(data=tips, x='day', y='total_bill')
plt.title('바이올린 플랏')
plt.show()
```

```python
# 카운트(개수)
sns.countplot(x=tips['day'])
```

```python
# 산점도(상관관계)
sns.relplot(data=tips, x='tip', y='total_bill')
```

```python
# 산점도 + 리니어(선)
sns.lmplot(data=tips, x='tip', y='total_bill')
```

```python
tips.corr()
```

```python
# 히트맵
sns.heatmap(tips.corr(), annot=True, linewidths=1)
```

```python
plt.bar(tips['day'], tips['total_bill'])
```

```python
# 바 그래프
sns.barplot(data=tips, x='day', y='total_bill', palette='rainbow') #palette : rainbow, coolwarm, 등등 여러가지 있음
```

```python
# catplot : 범주형 변수와 수치형 변수간의 관계
sns.catplot(data=tips, x='sex', y='total_bill', hue='smoker')
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', hue='smoker', kind='bar')
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', hue='smoker', kind='bar', col='time')
```

```python
# scatter
x = np.random.rand(50)
y = np.random.rand(50)
z = x*y*1000
plt.scatter(x, y, s=z, c=np.arange(50))
```

```python
sns.scatterplot(x, y, size=z, sizes=(z.min(), z.max()), hue=z, palette='coolwarm')
plt.show()
```

```python
# 파이차트
# 옵션 explode : 파이차트에서 한 조각이 튀어나오는 비율
labels = ['삼성', '애플', '화웨이', '샤오미', '기타']
values = [40, 30, 20, 7, 3]
explode = [0.1, 0, 0, 0, 0]
# startangle : 시작각도
# autopct : 퍼센트 표기
plt.pie(values, labels=labels, shadow=True, explode=explode, startangle=90, autopct='%.1f%%')
plt.show()
```

```python
# pairplot
# 그리드형태로 히스토그램과 분포도
sns.pairplot(tips, hue='size', palette='rainbow')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
# regression : 회귀선
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

```python
# hex 밀도
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
# kde : 등고선 형태
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```
