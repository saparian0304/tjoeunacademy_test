```python
# 2. 상위 5개 출력
print(data.head())
```

```python
# 3. 데이터의 행과 열 출력
print(data.shape)
```

```python
# 4. 데이터 프레임 확인
print(data.info())
```

```python
# 5. 전체 컬럼 목록 출력
print(data.columns)
```

```python
# 6. 6번째 컬럼 출력
print(data.columns[5])
```

```python
# 7. 6번째 컬럼의 타입 출력
print(data.iloc[:, 5].dtype)
```

```python
# 8. 6번째 컬럼의 3번째 값 출력
print(data.iloc[2,5])
```

```python
# 10. 각 수치형 변수의 분포(사분위, 평균, 표준편차, 최대, 최소) 확인
print(data.describe())
```

```python
# 11. 거주인구 컬럼값 출력
print(data['거주인구'])
```

```python
# 12. 평균 속도의 4분위수(IQR) 값 출력(3사분위값 - 1사분위값)
print(data['평균 속도'].describe().loc['75%'] - data['평균 속도'].describe().loc['25%'])
print(data['평균 속도'].quantile(0.75) - data['평균 속도'].quantile(0.25))
```

```python
# 13. 읍면동명 유일값 출력
print(data['읍면동명'].unique())
```

```python
# 14. 읍면동명이 연동인 데이터 상위 5개 출력
print(data[data['읍면동명'] == '연동'].head())
```

```python
# 15. 일자컬럼을 통해 년,월,일 컬럼 추가
data['일자'] = pd.to_datetime(data['일자'])
data['년'] = data['일자'].dt.year
data['월'] = data['일자'].dt.month
data['일'] = data['일자'].dt.day
print(data.head())
```

```python
# 16. 시도명과 읍면동명만 추출해서 새로운 데이터프레임(df2)로 생성
data2 = data[['시도명', '읍면동명']]
print(data2)
```

```python
# 17. 시도명이 제주시인 데이터 개수 출력
print(len(data[data['시도명']=='제주시']))
```

```python
# 18. 연동의 2018년 1월 데이터 출력
print(data[(data['읍면동명'] == '연동') & (data['년'] == 2018) & (data['월'] == 1)])
```

```python
# 19. 거주인구 순으로 정렬하고 인덱스 초기화
print(data.sort_values(by=['거주인구']).reset_index(drop=True))
```

```python
# 20. 읍면동명에 숫자 1이 들어가 있는 데이터 출력
print(data[data['읍면동명'].str.find('1') != -1])
```

```python
# 21. 거주인구 내림차순으로 정렬하고 인덱스 초기화
print(data.sort_values('거주인구', ascending=False).reset_index(drop=True))
```

```python
# 22. 읍면동명이 연동이거나 외도동인 데이터 출력
print(data[(data['읍면동명'] == '연동') | (data['읍면동명'] == '외도동')])
```

```python
# 23. 총 유동인구가 평균 유동인구보다 큰 데이터 출력
print(data[data['총 유동인구'] > data['총 유동인구'].mean()])
```

```python
# 24. id가 22457인 평균속도를 nan 값으로 변경
print(data[data['id'] == 22457])
import numpy as np
data.loc[data['id'] == 22457, '평균 속도'] = np.nan
print(data[data['id'] == 22457])
```

```python
# 25. 평균속도의 결측치 확인
print(data['평균 속도'].isnull().sum())
```

```python
# 26. 평균속도의 결측치를 평균속도의 평균으로 치환
data.loc[data['평균 속도'].isnull(), '평균 속도'] = data['평균 속도'].mean()
print(data['평균 속도'].isnull().sum())
```

```python
# 27. 읍면동명 데이터 중 읍 데이터의 갯수 출력
print(len(data[data['읍면동명'].str[-1] == '읍']))
print((data['읍면동명'].str.endswith('읍')).sum())
```

```python
# 28. 읍면동명이 3자인 데이터 갯수 출력
print(len(data[data['읍면동명'].str.len() == 3]))
```

```python
# 29. 시도명별 읍면동명 개수 출력
print(data.groupby(['시도명', '읍면동명']).size())
```

```python
data2 = pd.DataFrame(data.groupby(['시도명', '읍면동명']).size())
data2= data2.reset_index()
print(data2['시도명'].value_counts())
```

```python
# 30. 시도명별 읍면동명별 거주인구 평균
print(data.groupby(['시도명', '읍면동명'])['거주인구'].mean())
print(data.groupby(['시도명', '읍면동명']).mean()['거주인구'])
```

```python
# 31. 시도명별 읍면동명별 거주인구 평균, 최소, 최대 한번에 출력
# print(data.groupby(['시도명', '읍면동명'])['거주인구'].describe()[['mean', 'min', 'max']])
print(data.groupby(['시도명', '읍면동명'])['거주인구'].agg(['mean', 'max', 'min']))
```

```python
# 32. 제주시의 읍면동명별 거주인구 평균, 최소, 최대 출력
print(data[data['시도명']=='제주시'].groupby(['읍면동명'])['거주인구'].agg(['mean', 'min', 'max']))
```

```python
# 33. 시도명2 컬럼을 생성해 제주시는 0, 서귀포시는 1로 저장
condition = data['시도명'] == '제주시'
data.loc[condition, '시도명2'] = 0
data.loc[~condition, '시도명2'] = 1
print(data['시도명2'])
```

```python
print(data['시도명'].map({'제주시' : 0, '서귀포시' : 1}))
```

```python
# 34. 읍면동2 컬럼을 생성해 읍은 0, 면은 1, 동은 2로 저장
data['읍면동2'] = data['읍면동명'].str[-1].map({'읍':0, '면':1, '동':2})
print(data.loc[:,['읍면동명', '읍면동2']])

```

```python
# 35. 요일컬럼 추가
data['요일'] = data['일자'].dt.day_of_week
print(data['요일'].value_counts())
```
