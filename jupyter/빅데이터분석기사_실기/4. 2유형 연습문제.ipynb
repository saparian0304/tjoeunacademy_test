# 2유형 연습문제

Stellar Classification 데이터 세트

* stellar_y_train.csv : 은하여부 데이터(학습용)
* stellar_X_train.csv : 천체 관측 특성 데이터(학습용)
* stellar_X_test.csv : 천체 관측 특성 데이터(제출용)

* 제출 파일(csv)형식 : 컬럼명 : index(X_test의 인덱스), galaxy (은하 여부 예측 확률)

전체 컬럼

alpha : 적경각  
delta : 적위각  
u : 측광 시스템의 자외선 필터  
g : 측광 시스템의 녹색 필터  
r : 측광 시스템의 적색 필터  
i : 측광 시스템의 근적외선 필터  
z : 측광 시스템의 적외선 필터  
redshift : 파장 증가에 따른 적색편이 값  
run_ID : 특정 검색을 식별하는데 사용되는 실행번호  
return_ID : 이미지 처리방법을 지정하는 재실행번호  
cam_col : 실행 중 관측선을 식별하는 카메라 열  
field_ID : 각 구역을 식별하는 구역 변호  
plate : SDSS에서 각 plate를 식별하기 위한 ID  
galaxy : 천체 은하 여부 (은하 : 1, 은하아님 : 0)  

```python
import pandas as pd
x_train = pd.read_csv('data/stellar_X_train.csv')
y_train = pd.read_csv('data/stellar_y_train.csv')
x_test = pd.read_csv('data/stellar_X_test.csv')
# 데이터 살펴보기
# print(x_train.shape, y_train.shape, x_test.shape)
# print(x_train.describe())
# print(x_train['alpha'].value_counts())
# print(x_test['alpha'].value_counts())
# 라벨링 확인하기 - 없음
# 이상값, 결측치 확인하기 = u, g, z
# u
# print(x_train[(x_train['u'] < lower) | (x_train['u'] > upper)])
lower = x_train['u'].mean() - 3*x_train['u'].std()
upper = x_train['u'].mean() + 3*x_train['u'].std()
x_train.loc[(x_train['u'] < lower), 'u'] = lower
x_train.loc[(x_train['u'] > upper), 'u'] = upper

lower = x_test['u'].mean() - 3*x_test['u'].std()
upper = x_test['u'].mean() + 3*x_test['u'].std()
x_test.loc[(x_train['u'] < lower), 'u'] = lower
x_test.loc[(x_train['u'] > upper), 'u'] = upper
# print(x_train.describe())

# g
# print(x_train[(x_train['u'] < lower) | (x_train['u'] > upper)])
lower = x_train['g'].mean() - 3*x_train['g'].std()
upper = x_train['g'].mean() + 3*x_train['g'].std()
x_train.loc[(x_train['g'] < lower), 'g'] = lower
x_train.loc[(x_train['g'] > upper), 'g'] = upper

lower = x_test['g'].mean() - 3*x_test['g'].std()
upper = x_test['g'].mean() + 3*x_test['g'].std()
x_test.loc[(x_train['g'] < lower), 'g'] = lower
x_test.loc[(x_train['g'] > upper), 'g'] = upper
# print(x_train.describe())

# g
# print(x_train[(x_train['u'] < lower) | (x_train['u'] > upper)])
lower = x_train['z'].mean() - 3*x_train['z'].std()
upper = x_train['z'].mean() + 3*x_train['u'].std()
x_train.loc[(x_train['z'] < lower), 'z'] = lower
x_train.loc[(x_train['z'] > upper), 'z'] = upper

lower = x_test['z'].mean() - 3*x_test['z'].std()
upper = x_test['z'].mean() + 3*x_test['z'].std()
x_test.loc[(x_train['z'] < lower), 'z'] = lower
x_test.loc[(x_train['z'] > upper), 'z'] = upper
# rerun_ID 칼럼 삭제
x_train.drop(columns='rerun_ID', inplace=True)
x_test.drop(columns='rerun_ID', inplace=True)

# print(x_train.describe())
# 데이터 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train['run_ID'] = scaler.fit_transform(x_train[['run_ID']])
x_test['run_ID'] = scaler.transform(x_test[['run_ID']])

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train['plate'] = scaler.fit_transform(x_train[['plate']])
x_test['plate'] = scaler.transform(x_test[['plate']])
# print(x_train.describe())
# print(scaler.fit_transform(x_train[['run_ID']]))

# 데이터 split
from sklearn.model_selection import train_test_split
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(x_train, y_train, test_size=0.2)
# print(X_TRAIN.shape, X_TEST.shape, Y_TRAIN.shape, Y_TEST.shape)

# 모델 학습
# from xgboost import XGBClassifier
# model = XGBClassifier()
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_test_predicted = model.predict(x_test)

# 모델 평가
from sklearn.metrics import f1_score
# print(f1_score(Y_TEST, y_test_predicted))


x_test_index = pd.DataFrame(x_test.index)
y_test_predicted = pd.DataFrame(y_test_predicted)
# 제출
result = pd.concat([x_test_index, y_test_predicted], axis=1)
result.columns = ['index', 'galaxy']

result.to_csv('12345.csv', index=False)
# print(pd.read_csv('12345.csv'))
```
