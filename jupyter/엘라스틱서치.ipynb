```python
# requests, json으로 
_id = 1
for i in range(1,21) :
    iStart = (i-1)*1000 + 1
    iEnd = i*1000
    url = "http://openapi.seoul.go.kr:8088/6c6e685379616e6a3130325476486f67/json/TbPublicWifiInfo/{}/{}".format(iStart, iEnd)
    data = requests.get(url).json() # API 호출 응답 받은 결과
    
    for row in data['TbPublicWifiInfo']['row']:
        gu_nm = row['X_SWIFI_WRDOFC']
        place_nm = row['X_SWIFI_MAIN_NM']
        place_x = float(row['LAT'])
        place_y = float(row['LNT'])
        doc = {
            "gu_nm" : gu_nm,
            "place_nm" : place_nm,
            "instl_xy" : {
                "lat" : place_x,
                "lon" : place_y
            }
        }
        # Elasticsearch indexing
        res = requests.put("http://localhost:9200/seoul_wifi/_doc/{}".format(_id), data=json.dumps(doc), headers=header)
        _id += 1
    print("END", iStart, "~", iEnd)
print("END")
```

### bulk api

```python
from elasticsearch import Elasticsearch, helpers
es = Elasticsearch()
docs = []

for i in range(1,2): # range(1,17)
    iStart = (i-1)*1000 + 1
    iEnd = i*1000
    url = "http://openapi.seoul.go.kr:8088/6c6e685379616e6a3130325476486f67/json/TbPublicWifiInfo/{}/{}".format(iStart, iEnd)
    data = requests.get(url).json()
    
    for row in data['TbPublicWifiInfo']['row']:
        gu_nm = row['X_SWIFI_WRDOFC']
        place_nm = row['X_SWIFI_MAIN_NM']
        place_x = float(row['LAT'])
        place_y = float(row['LNT'])
        doc = {
            "_index" : "seoul_wifi2",
            "_source" : {
                "gu_nm" : gu_nm,
                "place_nm" : place_nm,
                "instl_xy" : {
                    "lat" : place_x,
                    "lon" : place_y
                }
            }   
        }
        docs.append(doc)
    print("END", iStart, "~", iEnd)

res = helpers.bulk(es, docs)
print("END")
```
