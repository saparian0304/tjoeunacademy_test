### pandas 내장 그래프
* plot() 메서드를 통해서 kind 옵션으로 그래프 종류 지정
* 그래프 종류
  - line : 선그래프 (기본값)
  - bar : 바그래프(수직)
  - barh : 바그래프(수평)
  - hist : 히스토그램
  - box : 박스플롯
  - kde : 커널 밀도 그래프
  - area : 면적
  - pie : 파이차트
  - scatter : 산점도
  - hexbin : 고밀도 산점도

```python
import pandas as pd
df = pd.read_excel('데이터수집/data/남북한발전전력량.xlsx')
df.head()
```

```python
df = pd.read_excel('데이터수집/data/시도별 전출입 인구수.xlsx')
df
```

```python
df.info()
```

```python
# 결측치 처리 (앞 데이터로 채우기)
df = df.fillna(method='ffill')
```

```python
# 서울에서 타지역으로 전입한 데이터만 추출
# df_seoul = 
# df[(df['전출지별'] =='서울특별시') * (df['전입지별'] != '전국') & (df['전입지별'] != '서울특별시')] # * 연산도 된다 (비트연산자 &와 동일하게 작동하는 것 같음)
df_seoul = df[(df['전출지별'] =='서울특별시') & (df['전입지별'] != '전국') & (df['전입지별'] != '서울특별시')]
```

```python
# 전출지별 컬럼 삭제
# df_seoul.drop(columns='전출지별')
df_seoul = df_seoul.drop('전출지별', axis=1)
```

```python
df_seoul = df_seoul.rename({'전입지별' : '전입지'}, axis=1)
```

```python
# 전입지 컬럼을 인덱스로 지정
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul
```

```python
# 서울에서 경기도로 전입한 데이터 선그래프
df_seoul.loc['경기도'].plot()
```

```python
df_seoul.loc['경기도'].index
df_seoul.loc['경기도'].values
```

```python
plt.plot(df_seoul.loc['경기도'].index, df_seoul.loc['경기도'].values)
```

```python
# 차트 스타일 지정
plt.style.use('ggplot') # 사용가능한 스타일은 plt.style.available에서 확인 가능
# 크기 지정
plt.figure(figsize=(15,4))
# x축 이름 회전
plt.xticks(rotation='vertical') # 회전각도 숫자 또는 'vertical', 'horizontal' 가능
# 마커
plt.plot(df_seoul.loc['경기도'], color='red', marker='s') # marker : ^, *, o, + 등등 help 확인
# 축이름
plt.xlabel('연도')
plt.ylabel('인구수')
# 제목
plt.title('서울->경기도')
# 범례 
plt.legend(labels=['서울->경기'])
plt.show()
```

```python
# 워닝메시지 무시
import warnings
warnings.filterwarnings("ignore")
```

```python
# 그래프 여러개
# 크기 지정
fig = plt.figure(figsize=(15,10)) # 큰 도화지
ax1 = fig.add_subplot(2, 1, 1) # 첫번째 그래프 (행, 열, 순서)
ax2 = fig.add_subplot(2, 1, 2) # 두번째 그래프 (행, 열, 순서)
# 마커
ax1.plot(df_seoul.loc['경기도'], color='red', marker='s') 
ax2.plot(df_seoul.loc['충청남도'], color='blue', marker='P')
# x축 이름 회전 (메서드가 달라짐)
ax1.set_xticklabels(df_seoul.loc['경기도'].index, rotation='vertical')
ax2.set_xticklabels(df_seoul.loc['충청남도'].index, rotation='vertical')

# y축의 범위(최소, 최대)
ax1.set_ylim(50000, 800000)
ax1.legend(labels=['서울->경기'], loc='upper left') # loc옵션 : 위치 지정
plt.show()
```

```python
# 하나의 axe객체에 여러개의 그래프 추가하기

# 크기 지정
fig = plt.figure(figsize=(15,4))
ax = fig.add_subplot(1,1,1)
# 마커
ax.plot(df_seoul.loc['전라남도'], color='red', marker='s')
ax.plot(df_seoul.loc['충청남도'], color='blue', marker='X')
ax.plot(df_seoul.loc['충청북도'], color='green', marker='o')

plt.show()
```

```python
# 면적
df2 = df_seoul.loc[['전라북도','전라남도','충청북도','충청남도']].T
df2.plot(kind='area', alpha=0.5)
```
