## SQLite
- 파이썬에 내장되어있음(핸드폰에도 SQLite가 내장되어있음)

```python
import sqlite3
conn = sqlite3.connect('test.sqlite')
```

```python
# sqlite는 기본 튜플형태로 리턴을 함
# 한 행을 딕셔너리로 출력하기 위해서 conn.row_factory() 함수를 오버라이드
# cur.description  : 각각 0번째의 요소가 테이블 컬럼명임
# row : 조회된 행 하나
def dictFactory(cursor, row):
    d = {}
    for idx, col in enumerate(cursor.description):
        d[col[0]] = row[idx]
    return d
conn.row_factory = dictFactory
```

```python
for i, letter in enumerate([['a','b','c'], ['a','b','c'], ['a','b','c']]):
    print(i, letter)
```

```python
cur = conn.cursor()
cur.executescript('''
drop table if exists test;
create table test(
    no integer primary key,
    name varchar,
    email varchar
)
''')
```

```python
cur.executescript('''
INSERT INTO test (no, name, email) VALUES (1, '홍길동', 'hong@gmail.com');
INSERT INTO test (no, name, email) VALUES (2, '김길동', 'kim@gmail.com');
INSERT INTO test (no, name, email) VALUES (3, '최길동', 'choi@gmail.com');
''')
conn.commit()
```

```python
cur.execute('SELECT * FROM test')
result = cur.fetchall()
result
```

```python
for r in result:
    print(r['name'])
```

```python
cur.close()
conn.close()
```

```python
# execute() : 하나의 실행문
# execure() : 2개 이상의 SQL문을 실행시킬때 사용한다
```
