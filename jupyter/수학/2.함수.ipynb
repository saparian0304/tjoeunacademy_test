```python
def H(x):
    if isinstance(x, np.ndarray):
        return np.where(x>=0, 1, 0)
    else:
        return 1 if x>=0 else 0
```

```python
H(-5), H(0), H(1.1), H(np.array([-1, 0, 1]))
```

```python
# 지수 함수
x = np.linspace(-2, 2, 100)
y = np.exp(x)
plt.plot(x, y)
plt.axvline(0, c='r', ls='--')
plt.axhline(1, c='r', ls='--')
plt.show()
```

```python
# 로그를 적용해도 함수의 최고점, 최저점의 위치는 변하지 않음
def f(x):
    return x**3 - 12*x + 20* np.sin(x) + 7
x = np.linspace(-4, 4, 300)
y = f(x)

plt.subplot(211)
plt.plot(x,y)
plt.yticks([0, 1, 5, 10])
plt.ylim(-2, 15)

plt.subplot(212)
plt.plot(x, np.log(y))

plt.show()
```

```python
# 시그모이드 함수
def sigmoid(x):
    return 1/(1+np.exp(-x))
x  = np.linspace(-10, 10, 100)
plt.plot(x, sigmoid(x))
plt.show()
```

```python
# 소프트플러스 함수 -> ReLU 함수와 비슷하나 x=0 근처에서 부드럽게 변함
def softplus(x):
    return np.log(1 + np.exp(x))

x = np.linspace(-10, 10, 100)
plt.plot(x, softplus(x))
plt.show()
```

```python
def f(x, y):
    return 2*x**2 + 6*x*y + 7*y**2 -26*x - 54*y + 107
```

```python
x = np.linspace(-3, 7, 100)
y = np.linspace(-3, 7, 100)
# z = f(x, y)
X, Y = np.meshgrid(x,y)
z = f(X, Y)
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, z)
plt.show()
```

```python
# 소프트맥스 함수
def softmax(x, w): # x : 입력값 벡터(1차원), w : 가중치 벡터
    e = np.exp(x*w)
    return e / e.sum()

x = [2, 1, 0.5]
y = softmax(x, np.ones(3))
y
```

```python
y.sum()
```

```python
# 가중치가 커지면 결과값의 차이가 더 커짐
softmax(x, np.ones(3)*4)
```

```python
# 함수의 이동
x = np.linspace(-10, 10, 100)
plt.plot(x, sigmoid(x), label='f(x)')
# plt.plot(x, sigmoid(x)-1, label='f(x)-1')
# plt.plot(x, sigmoid(x-5), label='f(x-5)')
plt.plot(x, sigmoid(x-5)-1, label='f(x-5)-1')
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, sigmoid(x), label='sigmoid')
plt.plot(x, 2*sigmoid(2*x)-1, label='tanh') # 탄젠트하이퍼볼릭 함수
plt.legend()
plt.show()
```
