```python
import numpy as np
```

```python
def f(x,y):
    return 2*x**2 + 6*x*y + 7*y**2 - 26*x - 54*y + 107
x = np.linspace(1, 16, 100)
y = np.linspace(-3, 6, 90)
X, Y = np.meshgrid(x, y)
Z = f(X,Y)

def gx(x, y):
    return 4*x + 6*y - 26
def gy(x, y):
    return 6*x + 14*y - 54

x2 = np.linspace(1, 16, 15)
y2 = np.linspace(-3, 6, 9)
X2, Y2 = np.meshgrid(x2, y2)
GX = gx(X2, Y2)
GY = gy(X2, Y2)

import matplotlib.pyplot as plt
# 컨투어
plt.contour(X, Y, Z, levels=np.logspace(0,3,10))
# 퀴버플롯
plt.quiver(X2, Y2, GX, GY)
plt.show()
```

```python
# 3d
ax = plt.subplot(projection='3d')
ax.plot_surface(X, Y, Z)
```

```python
def f(x):
    return (x-2)**2 + 2
x = np.linspace(-1, 4, 100)
plt.plot(x, f(x))
plt.ylim(0,10)
plt.show()
```

```python
# 도함수
def fd(x):
    return 2*x - 4

# 경사하강법
x = np.linspace(-1, 4, 100)
plt.plot(x, f(x), 'k--')

# step
size = 0.4

# 시작값
x1 = 0
plt.plot(x, fd(x1)*x + f(x1), 'b--')
print('첫번째 기울기값:', fd(x1))

x1 = x1 - size*fd(x1)
plt.plot(x, fd(x1)*(x-x1) + f(x1), 'r--')
print('두번째 기울기값:', fd(x1))

x1 = x1 - size*fd(x1)
plt.plot(x, fd(x1)*(x-x1) + f(x1), 'g--')
print('세번째 기울기값:', fd(x1))

x1 = x1 - size*fd(x1)
plt.plot(x, fd(x1)*(x-x1) + f(x1), 'y--')
print('네번째 기울기값:', fd(x1))

plt.ylim(0,10)
plt.show()
```
