```python
# 행렬 곱
A = np.array([[1,2,3], [4,5,6]])
B = np.array([[1,2], [3,4], [5,6]])
A.shape, B.shape
```

```python
# 행렬곱
A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])
C = np.array([[9, 8], [7,6]])
A @ B, B @ A # 교환법칙 성립 안함
```

```python
# 분배법칙
A@(B+C), A@B + A@C
```

```python
(A+B)@C, A@C + B@C
```

```python
(A+B).T, A.T + B.T
```

```python
(A@B).T, B.T @ A.T # 순서가 바뀌는 것 주의
```

```python
(A@B@C).T, C.T @ B.T @ A.T
```

```python
# 항등행렬의 곱셈
A = np.array([[1,2], [3,4]])
I = np.eye(2) # 2행2열짜리 항등행렬
I
```

```python
A@I, I@A
```
