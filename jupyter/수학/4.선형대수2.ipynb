```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 벡터의 평행이동
a1 = np.array([1,2])
plt.plot(a1)  # 파란선
plt.plot(a1+1) # 주황색
plt.ylim(-1, 4)
plt.xlim(-1, 4)
plt.show()
```

```python
# 놈과 a1의 내적
np.sqrt(a1.T@a1) # a1 제곱후 제곱근
```

```python
# 놈
np.linalg.norm(a1)
```

```python
plt.annotate('', a1, xytext=(0,0), arrowprops={'color':'gray', 'width':1})
plt.ylim(-1, 3)
plt.xlim(-1, 3)
plt.show()
```

```python
a2 = 2 * a1
a3 = -1 * a1

plt.annotate('', a1, xytext=(0,0), arrowprops={'color':'gray', 'width':1})
plt.annotate('', a2, xytext=(0,0), arrowprops={'color':'blue', 'width':1})
plt.annotate('', a3, xytext=(0,0), arrowprops={'color':'red', 'width':1})
plt.ylim(-4, 5)
plt.xlim(-4, 5)
plt.show()
```

```python
# 단위벡터
a = np.array([1, 0])
b = np.array([0, 1])
np.linalg.norm(a), np.linalg.norm(b)
```

```python
# 랭크
x1 = np.array([
    [1,3],
    [2,3]
])
x2 = np.array([
    [1,3,5],
    [2,3,7]
])
```

```python
# 기본값은 행랭크(row rank)
np.linalg.matrix_rank(x1), np.linalg.matrix_rank(x2)
```

```python
# 3행 = 1행*2
x3 = np.array([
    [1,3,5],
    [2,3,7],
    [2,6,10]
])
np.linalg.matrix_rank(x3) # 서로 독립 행벡터는 2개, 3개벡터 중에 종속관계가 하나 존재함
```

```python
A = np.array([[1,5,6], [2,6,8], [3,11,14], [1,4,5]])
B = np.array([[1,5,6], [2,6,8], [3,11,14], [1,4,8]])
```

```python
np.linalg.matrix_rank(A), np.linalg.matrix_rank(B) # A는 풀랭크가 아님
```

```python
# 풀랭크가 아니면 역행렬이 안구해지고, 회귀분석도 잘 안됨
np.linalg.inv(A)
```

```python
A = np.array([
    [1,5,6], 
    [2,6,8], 
    [3,11,5]
])
np.linalg.inv(A)
```

```python
# determinent (역행렬이 존재하는지 여부)
# 0이면 역행렬이 존재하지 않음
np.linalg.det(A)
```

```python
A = np.array([[1,5,6], [2,6,8], [0,0,0]])
np.linalg.det(A)
```

```python
np.linalg.inv(A)
```

```python
# 고유값과 고유벡터
A = np.array([[1,-2], [2,-3]])
# A = np.array([[2,4], [-1,-3]])
w, v = np.linalg.eig(A) # 고유값, 고유벡터 튜플형태로 리턴
```

```python
# 특이값과 특이벡터
A = np.array([[3,-1], [1,3], [1,1]])
A
```

```python
U, E, V = np.linalg.svd(A)
U, E, V
```

```python
# 대각행렬
np.diag(E, 1)[:, 1:]
```
