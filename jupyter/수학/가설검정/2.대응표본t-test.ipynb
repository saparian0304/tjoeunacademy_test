```python
import pandas as pd
import numpy as np
import scipy as sp
import scipy.stats
```

```python
df = pd.read_csv("../../데이터수집/data/다이어트약_효과검증.csv")
df
```

```python
before = df['다이어트전(kg)']
after = df['다이어트후(kg)']
before
```

```python
# 차이 정규성 검정
sp.stats.kstest(after-before, 'norm') # p-value가 0.05보다 작으므로 정규성
```

```python
sp.stats.ttest_rel(before, after) # pvalue가 0.05보다 작으므로 대립가설('차이가 있다') 채택
```
