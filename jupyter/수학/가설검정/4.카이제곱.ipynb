```python
import pandas as pd
import numpy as np
import scipy as sp
import scipy.stats
```

```python
df = pd.read_csv("../../데이터수집/data/htest05.csv")
df
```

```python
# 교차테이블(분할표)
pd.crosstab(df['smoke'], df['disease']).values # 교차테이블 만들기
```

```python
v = pd.crosstab(df['smoke'], df['disease']).values
sp.stats.chi2_contingency(v) # 통계량, pvalue, 자유도, 기대값
```

```python
# 성별에따른만족도.csv
df = pd.read_csv("../../데이터수집/data/성별에따른만족도.csv", encoding='ms949')
v = pd.crosstab(df['만족도'], df['성별'])
sp.stats.chi2_contingency(v)
```
